### 5.1 An√°lise do Target (`status_emprestimo`)

Nesta etapa, analisamos a distribui√ß√£o da vari√°vel alvo (`status_emprestimo`) para entender a propor√ß√£o de empr√©stimos aprovados e negados. Esse passo √© fundamental para identificar um poss√≠vel desbalanceamento entre as classes, o que pode influenciar as m√©tricas de avalia√ß√£o e a escolha dos algoritmos de machine learning.

Visualizaremos a distribui√ß√£o do target em:
1. **Tabela de Frequ√™ncia**: Para observar as quantidades absolutas e relativas de cada classe.
2. **Gr√°fico de Barras**: Para uma representa√ß√£o visual da distribui√ß√£o entre as classes.

```python
# Calculando a quantidade e a porcentagem de cada categoria do target 'status_emprestimo'
target_counts = df['status_emprestimo'].value_counts()
target_percentages = df['status_emprestimo'].value_counts(normalize=True) * 100

# Exibindo a quantidade e porcentagem
print("Distribui√ß√£o de 'status_emprestimo' (quantidade e porcentagem):")
for status, count in target_counts.items():
    print(f"Status {status}: {count} ({target_percentages[status]:.2f}%)")

# Plotando a distribui√ß√£o
plt.figure(figsize=(6, 4))
bars = target_counts.plot(kind='bar', color=['#4f81bd', '#aec6cf'])  # Tons de azul

# Adicionando porcentagens no topo de cada barra
for bar, percentage in zip(bars.patches, target_percentages):
    bars.annotate(f"{percentage:.2f}%", 
                  (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                  ha='center', va='bottom')

plt.title("Distribui√ß√£o do Target 'status_emprestimo'")
plt.xlabel("Status do Empr√©stimo")
plt.ylabel("Quantidade")
plt.xticks(rotation=0)
plt.show()
```

---
### üí° **Observa√ß√µes**:

- A an√°lise da vari√°vel `status_emprestimo` revela um **desbalanceamento entre as classes**:
  - **Rejei√ß√µes** (status = 0): Representam 85.76% dos dados, indicando uma predomin√¢ncia de empr√©stimos negados.
  - **Aprova√ß√µes** (status = 1): Representam apenas 14.24% dos dados, mostrando uma menor propor√ß√£o de empr√©stimos aprovados.

- Esse desbalanceamento sugere que medidas espec√≠ficas para lidar com classes desbalanceadas, como ajustes nas m√©tricas de avalia√ß√£o ou t√©cnicas de amostragem, podem ser necess√°rias durante a modelagem para garantir que o modelo tenha um bom desempenho em ambas as classes.

## 5.4 An√°lise de Outliers üîç <a id="5-4-analise-de-outliers"></a>

Nesta se√ß√£o, realizaremos uma an√°lise detalhada de outliers presentes nas vari√°veis quantitativas. Outliers s√£o valores que se afastam significativamente da distribui√ß√£o central dos dados e podem impactar o desempenho dos modelos de machine learning, dependendo de sua natureza e causa.

As etapas desta an√°lise incluem:
1. Identifica√ß√£o de outliers nas vari√°veis quantitativas, utilizando m√©todos estat√≠sticos como o **Intervalo Interquartil (IQR)**.
2. Visualiza√ß√£o dos outliers com **boxplots** para cada vari√°vel, o que nos permitir√° avaliar a extens√£o dos desvios e poss√≠veis impactos nos dados.
3. Discuss√£o sobre o tratamento adequado desses outliers para a etapa de modelagem, considerando se eles devem ser mantidos ou tratados de forma especial.

Essas etapas s√£o essenciais para garantir que o modelo de machine learning possa generalizar bem e evitar vi√©s causado por valores extremos no dataset.

```python
# Inicializando um dicion√°rio para armazenar a quantidade de outliers para cada vari√°vel
outliers_info = {}

# Calculando os limites e contagem de outliers para cada vari√°vel quantitativa
for coluna in colunas_quantitativas:
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    # Contando os outliers abaixo e acima dos limites
    outliers_abaixo = df[df[coluna] < limite_inferior].shape[0]
    outliers_acima = df[df[coluna] > limite_superior].shape[0]
    total_outliers = outliers_abaixo + outliers_acima

    # Armazenando os resultados no dicion√°rio
    outliers_info[coluna] = {
        'Outliers Abaixo': outliers_abaixo,
        'Outliers Acima': outliers_acima,
        'Total de Outliers': total_outliers
    }

# Convertendo o dicion√°rio para um DataFrame para melhor visualiza√ß√£o
tabela_outliers = pd.DataFrame(outliers_info).T

# Exibindo a tabela de frequ√™ncia de outliers
tabela_outliers
```

```python
# Calculando os limites inferior e superior de outliers para cada vari√°vel quantitativa
faixas_outliers = {}

for coluna in colunas_quantitativas:
    Q1 = df[coluna].quantile(0.25)
    Q3 = df[coluna].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    faixas_outliers[coluna] = {
        'Q1': Q1,
        'Q3': Q3,
        'Limite Inferior': lower_bound,
        'Limite Superior': upper_bound,
        'Min': df[coluna].min(),
        'Max': df[coluna].max()
    }

# Convertendo para DataFrame para visualiza√ß√£o
df_faixas_outliers = pd.DataFrame(faixas_outliers).T

# Exibindo a faixa de valores dos outliers
display(df_faixas_outliers)
```

```python
# Definindo limites plaus√≠veis para cada vari√°vel
limites_extremos = {
    'idade_pessoa': 80,  # Definido como um limite plaus√≠vel m√°ximo
    'tempo_emprego_pessoa': 40,  # Tempo de emprego acima de 40 anos pode ser considerado extremo
    'renda_pessoa': 1_000_000,  # Limite de 1 milh√£o para renda anual
    'tempo_historico_credito_pessoa': 30  # Tempo de hist√≥rico de cr√©dito acima de 30 anos √© considerado extremo
}

# Identificando registros com valores extremamente altos em cada vari√°vel
extremos = {}
for coluna, limite in limites_extremos.items():
    extremos[coluna] = df[df[coluna] > limite][[coluna, 'status_emprestimo']]
    
# Exibindo as observa√ß√µes com valores extremos
for coluna, dados_extremos in extremos.items():
    print(f"Valores extremos para {coluna}:")
    display(dados_extremos)
    print("\n" + "-"*50 + "\n")
```

- **idade_pessoa**:

Assim, ser√° adequado remover valores de **idade_pessoa** acima de 80 anos para evitar distor√ß√µes na an√°lise.

- **tempo_emprego_pessoa**:

Contudo, para manter a coer√™ncia nos dados, recomendaremos limitar o **tempo_emprego_pessoa** a 40 anos, removendo registros acima desse valor.

- **renda_pessoa**:
  - H√° rendas extremamente altas no dataset, com valores chegando a **1.9 milh√µes**. Dependendo do foco do modelo, esses valores podem ser mantidos se a modelagem precisar considerar solicitantes de alta renda. No entanto, esses valores elevados podem ser removidos para evitar que os dados sejam influenciados por extremos, caso o objetivo seja trabalhar com uma faixa de renda mais t√≠pica. ### 5.5 An√°lise de Correla√ß√£o üîó <a id="5-5-analise-de-correlacao"></a>

Nesta se√ß√£o, exploramos as correla√ß√µes entre as vari√°veis num√©ricas do dataset para identificar rela√ß√µes significativas e avaliar a possibilidade de multicolinearidade. Compreender a correla√ß√£o entre as vari√°veis √© fundamental para a constru√ß√£o de modelos de machine learning, pois vari√°veis altamente correlacionadas podem impactar a interpreta√ß√£o dos resultados e a efic√°cia do modelo.

Os objetivos principais desta an√°lise incluem:

1. **Identifica√ß√£o de Correla√ß√µes Fortes**:
   - Detectar vari√°veis que possuem alta correla√ß√£o com o target (`status_emprestimo`), pois elas podem ter maior import√¢ncia preditiva.
   - Identificar correla√ß√µes significativas entre as vari√°veis independentes, o que pode levar √† multicolinearidade e afetar a robustez do modelo.

2. **Visualiza√ß√£o da Matriz de Correla√ß√£o**:
   - Utilizaremos um mapa de calor (heatmap) para visualizar as correla√ß√µes de maneira intuitiva, destacando as vari√°veis com maiores coeficientes de correla√ß√£o.
   
3. **Interpreta√ß√£o**:
   - Vari√°veis com correla√ß√£o forte e direta com o target ser√£o consideradas priorit√°rias no processo de sele√ß√£o de features.

Essa an√°lise ajudar√° a identificar vari√°veis potencialmente mais relevantes para o modelo e facilitar√° a cria√ß√£o de um conjunto de dados mais robusto e bem estruturado.

```python
# Calculando a matriz de correla√ß√£o
correlacao = df.corr()

# Configura√ß√£o do tamanho da figura
plt.figure(figsize=(12, 8))

# Plotando o mapa de calor
sns.heatmap(correlacao, annot=True, fmt=".2f", cmap="Blues", cbar=True, square=True,
            linewidths=0.5, linecolor='black')

# T√≠tulos e exibi√ß√£o
plt.title("Mapa de Calor da Matriz de Correla√ß√£o")
plt.show()
```
